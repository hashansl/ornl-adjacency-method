```python
overdose_df = pd.read_excel('/home/h6x/git_projects/ornl-svi-data-processing/raw_data/HepVu_County_Opioid_Indicators_05DEC22.xlsx',dtype={'GEO ID': str})
```

```python
overdose_df = overdose_df[['GEO ID','County Name','Narcotic Overdose Mortality Rate 2018']]
```

```python
overdose_df = overdose_df.rename(columns={'Narcotic Overdose Mortality Rate 2018': 'NOD_Rate','GEO ID':'STCNTY'})
```

```python
overdose_df.head(3)
```

```python
# Calculate the min and max of the column
min_value = overdose_df['NOD_Rate'].min()
max_value = overdose_df['NOD_Rate'].max()
```

```python
min_value, max_value
```

```python
# Assuming negative overdose rate means this data is missing
len(overdose_df[overdose_df['NOD_Rate'] < 0])
```

```python
# lets not consider the missing data
overdose_df = overdose_df[overdose_df['NOD_Rate'] >= 0]
```

```python
overdose_df.shape
```

```python
overdose_df['OD_class'] = pd.qcut(overdose_df['NOD_Rate'], q=[0,0.90, 1], labels=['0', '1'])
```

```python
overdose_df = overdose_df.reset_index(drop=True)
```

```python
overdose_df['STCNTY'] = overdose_df['STCNTY'].apply(lambda x: '0' + x if len(x) == 4 else x)
```

```python
overdose_df.sort_values(by=['NOD_Rate'], ascending=False).head(5)
```

```python
# make the STCNTY column STR
df['STCNTY'] = df['STCNTY'].apply(lambda x: str(x))
```

```python
# For STCNTY column, we need to add a leading zero if the length of the string is 4
df['STCNTY'] = df['STCNTY'].apply(lambda x: '0' + x if len(x) == 4 else x)
```

```python
# add NOD_Rate and OD_class to the df based on the STCNTY
merged_df = pd.merge(df, overdose_df[['STCNTY','NOD_Rate','OD_class']], on='STCNTY', how='left')
```

```python
for var in variables:

    for i in range(0, 11):
        df_var = merged_df[merged_df['Variable'] == var]
        count = df_var[df_var[f'H1_count'] == i].shape[0]
        print(f'Variable {var} with H1_count {i} = {count}')

        high_od_count_h1 = df_var[(df_var[f'H1_count'] == i) & (df_var[f'OD_class'] == '1')].shape[0]
        print(f'OD_class 1 count: {high_od_count_h1}')

        low_od_count_h1 = df_var[(df_var[f'H1_count'] == i) & (df_var[f'OD_class'] == '0')].shape[0]
        print(f'OD_class 0 count: {low_od_count_h1}')

    df_temp = merged_df[merged_df['Variable'] == var]
    more_count = df_temp[df_temp['H1_count'] > 10].shape[0]
    print(f'Variable {var} with H1_count > 10 = {more_count}')

    high_od_count = df_temp[(df_temp[f'H1_count'] > 10) & (df_temp[f'OD_class'] == '1')].shape[0]
    print(f'OD_class 1 count: {high_od_count}')

    low_od_count = df_temp[(df_temp[f'H1_count'] > 10 ) & (df_temp[f'OD_class'] == '0')].shape[0]
    print(f'OD_class 0 count: {low_od_count}')

    print('----------------------')

    # break
```
